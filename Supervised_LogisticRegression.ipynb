```python
# Logistic Regression:
# Key Concept:
# Logistic Regression is used for binary classification (two classes, e.g., spam or not spam).
# Logistic regression predicts the probability that an instance belongs to a certain class.
```

```python
# How it works:
# Logistic regression uses the logistic function (also called the sigmoid function) to output a value between 0 and 1. 
# This represents the probability of belonging to the "positive" class (e.g., spam or disease).
# If the output is greater than 0.5, the instance is classified as class 1 (e.g., spam).
# If the output is less than 0.5, it’s classified as class 0 (e.g., not spam).
```

```python
# 🧠 What is Logistic Regression?
# Analogy: Think of a yes/no decision — like deciding whether to bring an umbrella.

'Let’s say you look outside, see some clouds, check the wind, feel the humidity, and then decide:'
"Should I carry an umbrella or not?"

#This is a binary decision — similar to what logistic regression does:
'''Inputs: cloud level, wind, humidity, etc.
Output: probability of rain (between 0 and 1)
If probability > 0.5 → take umbrella (YES / 1)
If probability ≤ 0.5 → don’t take umbrella (NO / 0)'''

# This is exactly how logistic regression works:
# It takes multiple features, combines them, and uses a special formula (sigmoid function) to output a probability.
# If that probability is high, it chooses one class; if low, it chooses another.
```

```python
# 1. Importing necessary libraries
from sklearn.model_selection import train_test_split   # Tools to split data into training/testing sets
from sklearn.linear_model import LogisticRegression    # The LogisticRegression model
from sklearn.datasets import load_iris                 # A sample dataset (Iris flower dataset 🌸)
from sklearn.metrics import accuracy_score             # A function to measure accuracy
```

```python
# 2. Load a simple dataset (Iris dataset for this example)
data = load_iris()
X = data.data
y = (data.target == 0).astype(int)  # Convert to binary classification (class 0 vs others)
```

```python
'''👉 இது Iris dataset-அ load பண்ணுது. இது ஒரு famous dataset – mostly machine learning-ல beginners use பண்ணுவாங்க.
Iris dataset-ல 3 types of flowers இருக்குது: Setosa, Versicolor, Virginica.'''


'''👉 இங்க X என்னன்னா, features.
அதாவது, ஒவ்வொரு flower-க்கும் என்ன என்ன measure பண்ணிருக்காங்க (like sepal length, petal width etc.), அதெல்லாம் X-ல store ஆகும்.'''


'''👉 இதுதான் little tricky but easy:
data.target = original label (0, 1, 2) → 0 = Setosa, 1 = Versicolor, 2 = Virginica.
data.target == 0 → இது ஒரு boolean array return பண்ணும்: Setosa-னா True, இல்லனா False.
.astype(int) → True-ன 1, False-ன 0 ஆக convert ஆகும்.

So finalா y-ல இருக்குது: 👉 Setosa-னா 1, இல்லனா 0.
I.e., இப்போ y ஒரு binary classification label ஆயிடுச்சு.

Summary in simple terms:
Iris dataset எடுத்துருக்கோம், அந்த data-வ features (X) and "Setosa-தானா இல்லையா?" (y) split பண்ணி எடுத்துருக்கோம்.

y-ல, Setosa-னா 1, இல்லனா 0.
இதுக்கு later use பண்ணி binary classifier train பண்ணலாம்.'''
```

```python
'''📌 This loads a famous dataset with 3 types of flowers (Iris setosa, versicolor, virginica).
But logistic regression works best for binary outcomes.
So we convert it into a Yes/No problem:
Is the flower Setosa (0) → YES (1)
Is it not Setosa → NO (0)
✅ This is now a binary classification problem.'''
```

```python
# 3. Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
'''இந்த line-ல நாம data-வை 2 parts-ஆ split பண்ணுறோம் — training set & testing set.
**X**னு இருக்கு — அதுதான் independent variables (features)
**y**ன்னா dependent variable (target)

✅ train_test_split(X, y, test_size=0.3)
இது சொல்லுது:
👉 70% data training-க்கு use பண்ணு
👉 30% data testing-க்கு வைக்க (that's the test_size=0.3)

✅ random_state=42
இதுல ஏதாவது randomness இருக்காம, எப்போதும் ஒன்னா reproducible result வரட்டும்-nu fix பண்ணறது. (like setting a seed)

✅ Final output:
X_train, y_train → model கற்றுக்க வேண்டியது
X_test, y_test → model-ஐ evaluate பண்ண test data'''
```

```python
'''| Part                 | Meaning |
|------                   |---------|
| `X`                     | Features (independent variables) — input data for prediction |
| `y`                     | Target (dependent variable) — the value you want to predict |
| `test_size=0.3`         | 30% of the data is reserved for testing, and 70% is used for training |
| `random_state=42`       | Sets a fixed random seed so that the split is **reproducible** every time you run the code |
| `X_train`               | Training input features |
| `X_test`                | Testing input features |
| `y_train`               | Training labels (what the model learns from) |
| `y_test`                | Testing labels (what we compare the predictions to) |'''

'''📌 Split the dataset:
70% for training
30% for testing
This helps us see how well the model can generalize to new, unseen data.'''
```

```python
# 4. Initialize the Logistic Regression model
# 📌 Create a logistic regression model object.

model = LogisticRegression()        
```

```python
# 5. Train the model
#📌 This is where the learning happens.The model finds the best weights (like deciding how important each feature is).    

model.fit(X_train, y_train)  
```

```python
'''👉 model.fit() — நம்ம model-ஐ train பண்ணுறோம் using training data.
X_train → input features (like study material)
y_train → correct answers (target/output)

So, **model.fit(X_train, y_train)**ன்னா:
➡️ Model-க்கு questions (X_train) & correct answers (y_train) கொடுத்து,
learning process பண்ணுறோம்.
It’s like coaching class-ல model padichu exam-க்கு prepare ஆகற மாதிரி.

📚 Simple Analogy:
Imagine model = student
X_train = questions
y_train = correct answers

fit() = padikka vechiradhu!

So, இவங்க ippo padichuttu ready for exam (predict)!'''
```

```python
# 6. Make predictions
#📌 The model uses the test data to make predictions:Is this flower Setosa (1) or not (0)?

y_pred = model.predict(X_test)
```

```python
'''👉 model.predict(X_test)
— அதாவது, நாம train பண்ணிய model-ஓட help-ல, unseen test data-கு (i.e., X_test) prediction பண்ணுறோம்.
— இவங்க output-ஐ y_predனு save பண்ணுறோம்.

🧠 So basically:
X_test → New input data (features only)
model.predict() → இதுல model use பண்ணி result (prediction) guess பண்ணும்
y_pred → Model guess பண்ணி கொடுத்த output values (prediction results)

🎯 Simple Analogy:
Namma model-க்கு training time-ல class-la padichathu X_train, y_train.
Ippo exam-la X_test-ல irukkura questions kudukkumbothu,
model-oda brain use பண்ணி answer solradhu dhaan y_pred!'''
```

```python
# 7. Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
'''இந்த code-ல நாம model-oda performance check பண்ணுறோம் — specifically accuracy.
y_test → Actual answers (exam answer key)
y_pred → Model guess பண்ணி சொன்ன answers

🔍 Line by line:
👉 accuracy_score(y_test, y_pred)
அது check பண்ணும்:
Model எத்தனை correct answers சொன்னது?
(எதுக்கு எவ்வளவு match ஆச்சு-nu)

👉 accuracy * 100
Percentage-ஆ convert பண்ணுறோம்.
Ex: 0.92 → 92%

👉 print(f"...")
Print பண்ணுறோம் with 2 decimal points like:
Accuracy: 92.35%

🎯 Analogy:
y_test = answer key
y_pred = student answer sheet

So, accuracy_score = paper correct pannra teacher!
எவ்வளவு correct answer-nu mark பண்ணிட்டு final score கொடுக்குறாங்க 😄'''
```
