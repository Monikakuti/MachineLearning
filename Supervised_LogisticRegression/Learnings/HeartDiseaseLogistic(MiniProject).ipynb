```python
# Step 6: Define features and target
X = df.drop("condition", axis=1)
y = df["condition"]
```

```python
# Step 7: Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Step 8: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Step 9: Train model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
# Step 10: Evaluate
y_pred = model.predict(X_test)
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Step 11: Manual Prediction Input
print("\n🧪 Manual Prediction")
print("Please enter the following health values:")
```

```python
# Features for prediction
input_features = [
    "age", "sex", "cp", "trestbps", "chol", "fbs", "restecg",
    "thalach", "exang", "oldpeak", "slope", "ca", "thal"
]

# Collect user inputs
user_data = []
for feature in input_features:
    value = float(input(f"Enter {feature}: "))
    user_data.append(value)
```

```python
# Convert to scaled format
user_df = pd.DataFrame([user_data], columns=input_features)
user_data_scaled = scaler.transform(user_df)

user_prediction = model.predict(user_data_scaled)[0]

# Output result
result = "likely to have Heart Disease 💔" if user_prediction == 1 else "not likely to have Heart Disease 💖"
print(f"\n🔍 Prediction Result: The person is {result}")
```
